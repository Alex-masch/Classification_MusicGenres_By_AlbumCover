##Visualizing data

```python
#setting dir names for matplot
hiphop_names = os.listdir("/content/hiphop")
metal_names = os.listdir("/content/metal")
classical_names = os.listdir("/content/classical")
country_names = os.listdir("/content/country")
edm_dance_names = os.listdir("/content/edm_dance")
jazz_names = os.listdir("/content/jazz")
latin_names = os.listdir("/content/latin")
pop_names = os.listdir("/content/pop")
punk_names = os.listdir("/content/punk")
rock_names = os.listdir("/content/rock")
soul_names = os.listdir("/content/soul")
```

```python
# Define matplot parameters to see how the cover art look like

%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Parameters for our graph; we'll output images in a 4x4 configuration
nrows = 4
ncols = 4

# Index for iterating over images
pic_index = 0
```

```python
# Set up matplotlib fig, and size it to fit 4x4 pics of spotify cover art
fig = plt.gcf()
fig.set_size_inches(ncols * 4, nrows * 11)

pic_index += 22
next_classical_pix = [os.path.join(classical_dir, fname) 
                #for fname in classical_names[pic_index-22:pic_index]]
next_metal_pix = [os.path.join(metal_dir, fname) 
                #for fname in metal_names[pic_index-22:pic_index]]
next_hiphop_pix = [os.path.join(hiphop_dir, fname) 
                #for fname in hiphop_names[pic_index-22:pic_index]]
next_country_pix = [os.path.join(country_dir, fname) 
                #for fname in country_names[pic_index-22:pic_index]]
next_edm_dance_pix = [os.path.join(edm_dance_dir, fname) 
                #for fname in edm_dance_names[pic_index-22:pic_index]]
next_jazz_pix = [os.path.join(jazz_dir, fname) 
                #for fname in jazz_names[pic_index-22:pic_index]]
next_latin_pix = [os.path.join(latin_dir, fname) 
                for fname in latin_names[pic_index-8:pic_index]]
next_pop_pix = [os.path.join(pop_dir, fname) 
                #for fname in pop_names[pic_index-22:pic_index]]
next_punk_pix = [os.path.join(punk_dir, fname) 
                for fname in punk_names[pic_index-8:pic_index]]
next_rock_pix = [os.path.join(rock_dir, fname) 
                #for fname in rock_names[pic_index-22:pic_index]]
next_soul_pix = [os.path.join(soul_dir, fname) 
                #for fname in soul_names[pic_index-22:pic_index]]

for i, img_path in enumerate(next_latin_pix+next_punk_pix):
  # Set up subplot; subplot indices start at 1
  sp = plt.subplot(nrows, ncols, i + 1)
  sp.axis('Off') # Don't show axes (or gridlines)

  img = mpimg.imread(img_path)
  plt.imshow(img)

plt.show()
```
